**1. Посчитайте, сколько мужчин и женщин (признак *sex*) представлено в этом датасете**

**4-5. Рассчитайте среднее значение и среднеквадратичное отклонение возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год**

**7. Выведите статистику возраста для каждой расы (признак *race*) и каждого пола. Используйте *groupby* и *describe*. Найдите таким образом максимальный возраст мужчин расы *Asian-Pac-Islander*.**

```python
# Группируем по двум колонкам и берем статистику возраста
age_by_race_sex_statistics = data.groupby(['race', 'sex'])['age'].describe()
age_by_race_sex_statistics
```

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
men_only = data[data['sex'] == 'Male'].copy()

#Создаем признак через лямбда-функцию
men_only['is_married'] = men_only['marital-status'].apply(lambda x: 'Married' if x.startswith('Married') else 'Single')

wealth_share = men_only.groupby('is_married')['salary'].value_counts(normalize=True)
wealth_share
```

```python
married_share = wealth_share['Married', '>50K']
single_share = wealth_share['Single', '>50K']

if married_share > single_share:
    print("More rich married men")
else:
    print("More rich single men")
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
# Максимум часов в неделю
max_hours = data['hours-per-week'].max()
print(f"Max hours per week: {max_hours}")
```

```python
# Выбираем работяг, которые работают максимум часов в неделю
workers_max_hours = data.loc[data['hours-per-week'] == max_hours]

# Считаем количесство бедолаг
count_workers_max_hours = len(workers_max_hours)
print(f"Workers count who work max hours per week: {count_workers_max_hours}")
```

```python
# Доля богатых (>50K) бедолаг
rich_workers_share = workers_max_hours['salary'].value_counts(normalize=True)
print(rich_workers_share * 100)
```

**11.Сгруппируйте людей по возрастным группам *young*, *adult*, *retiree*, где:**
* *young* соответствует 16-35 лет
* *adult* - 35-70 лет
* *retiree* - 70-100 лет

**Проставьте название соответсвтуещей группы для каждого человека в новой колонке AgeGroup**

```python
# (16, 34] - young
bins = [16, 34, 70, 100]
labels = ['young', 'adult', 'retiree']

# include_lowest - чтобы включить 16-летних
data['AgeGroup'] = pd.cut(data['age'], bins=bins, labels=labels, include_lowest=True)

data[['age', 'AgeGroup']]
```

**12-13. Определите количество зарабатывающих >50K в каждой из возрастных групп (колонка AgeGroup), а также выведите название возрастной группы, в которой чаще зарабатывают больше 50К (>50K)**

```python
rich = data[data['salary'] == '>50K']
rich_age_groups = rich.groupby('AgeGroup', observed=False).size().reset_index(name='count')

print(f"Age groups that make >50K:\n")
rich_age_groups
```

```python
name_rich_age_group = data.groupby('AgeGroup', observed=False)['salary'].value_counts(normalize=True)
max_rich = name_rich_age_group.loc[:,'>50K'].idxmax()
max_rich
```

**14. Сгруппируйте людей по типу занятости (колонка occupation) и определите количество людей в каждой группе. После чего напишите функциюю фильтрации filter_func, которая будет возвращать только те группы, в которых средний возраст (колонка age) не больше 40 и в которых все работники отрабатывают более 5 часов в неделю (колонка hours-per-week)**

```python
occupation_groups = data.groupby('occupation')
occupation_groups.size().reset_index(name='count')
```

```python
def filter_func(group_to_filter):
    avg_age = group_to_filter['age'].mean() <= 40
    all_work_enough = (group_to_filter['hours-per-week'] > 5).all()

    return avg_age and all_work_enough

filtered_data = occupation_groups.filter(filter_func)
filtered_data['occupation'].unique()
```
